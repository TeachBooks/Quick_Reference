| Type | Size | $h$ (mm) | $b$ (mm) | $M_{z;el;d}$ S235 (kNm) | $M_{z;el;d}$ S355 (kNm) |
|------|------|--------|--------|--------------------|--------------------|
| HE A | 100 - 1000 | 96 - 990 | 100 - 300 | 17,1 - 2629 | 25,8 - 3972 |

| HE M | 100 - 1000 | 120 – 1008 | 106 - 302 | 44,8 - 3368 | 67,6 – 5088 |


Applications: H-sections are useful for compression elements (columns) or beams where the structural height has to be limited. The increase in section modulus is achieved by using thicker flanges. Due to their large width the H-sections retain their shape relatively well.
